# 1. Total Profit by Item Type (Bar Chart)

```python
item_profit = df.groupby('Item Type')['Total Profit'].sum().reset_index()

# Sort by Total Profit
item_profit = item_profit.sort_values(by='Total Profit', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Total Profit', y='Item Type', data=item_profit, palette='coolwarm')
plt.title('Total Profit by Item Type')
plt.xlabel('Total Profit')
plt.ylabel('Item Type')
plt.tight_layout()
```

# 2. Units Sold by Sales Channel (Pie Chart)

```python
sales_channel_units = df.groupby('Sales Channel')['Units Sold'].sum()

# Plotting Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(sales_channel_units, labels=sales_channel_units.index, autopct='%1.1f%%', colors=['#66b3ff','#99ff99'], startangle=90, explode=(0.05, 0))
plt.title('Units Sold by Sales Channel')
plt.tight_layout()

# Show plot
plt.show()
```
